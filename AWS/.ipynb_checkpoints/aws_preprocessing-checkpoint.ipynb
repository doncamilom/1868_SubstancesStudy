## For this challenge we use MAPE as the evaluation metric:

\begin{equation}
MAPE = \frac{100}{n}* \Sigma_{i=1}^{n}\frac{ | \hat{y_i} - y_i|}{y_i}
\end{equation}


```python
def MAPE(y_true,y_pred):
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    
    return np.mean(np.abs((y_true - y_pred)/y_true)) * 100
```

#  Let's start doing some EDA and feature generation

```python
def plotFeat_seg(df,col,lim=15,rot=90):
    """ Make a barplot of normalized counts for train and test set for the same variable"""

    grp = df.groupby(['train',col]).id_cli.count()
    grp.loc[0] = grp/grp[0].sum()
    grp.loc[1] = grp/grp[1].sum()
    grp = grp.reset_index()
    grp[col] = grp[col].astype(str)

    fig,ax = plt.subplots(figsize=(15,2.5))
    
    if grp.shape[0] > lim:  # If too many categs, pick 15 most common
        grp = grp.sort_values('id_cli',ascending=False).iloc[:lim]
        ax.set_xticks(grp.index)
        ax.set_xticklabels(grp.index,rotation=rot)    

    ax.set_xticklabels([],rotation=rot)
    sns.barplot(data=grp, x=col,y='id_cli',hue='train')

    plt.show()
```

```python
# Let's start merging some values.
df.nivel_academico = df.nivel_academico.replace(['NO INFORMA','SIN INFORMACION'],'N/I')

# There is some ordinal structure in the nivel_academico col. 
cats = ['N/I', 'NINGUNO','PRIMARIO', 'BACHILLER','TECNOLOGO', 
        'UNIVERSITARIO','ESPECIALIZACION' ]

mapp = {cat:i for i,cat in enumerate(cats)}

df['academ_order'] = df.nivel_academico.replace(mapp)

# Produce frequency encodings
for col in var:  # All of these are str variables
    grp = (df.groupby(col)
           .id_cli.count().reset_index()
           .rename(columns={'id_cli':col+'_freq'}))
    
    df = df.merge(grp,how='right',on=col)
```

## Ocupacion y profesion

```python
df.ocupacion.value_counts().head(60)

# Some profesions are listed here as well
# Let's concatenate ocupacion - profesion and run next code with that

df.loc[:,'prof_ocup'] = (df.profesion.astype(str) + ' ' + df.ocupacion.astype(str)).str.upper()
df.head()
```

```python
print(df.profesion.value_counts().head(5))

replace = {'EDUCACION':['PROFES','EDUCACION'],
           'SALUD':['MEDIC','PSICOLOGIA','ENFERM','BACTERIOLOGIA','VETERINARIA','ODONT',
              'SALUD','NUTRICION Y DIETETICA','FARMA'],
           'ECONOM':['CONTA','ECONOMIA','COMERC','MERCAD'],
           'CONSTRU':['ARQUITE','CONSTRU','GEOLOGIA','MINAS'],
           'SOCIAL':['DERECHO','SECRETARIADO','SOCIAL','SOCIOL','HISTORIADOR',
              'BIBLIOTECARIOS','FILOSOFIA Y LETRAS','ESCRITORES','SACERDOTE','FILOSOFOS','RELIGIOSA'],
           'AGRO':['AGRO','BIOLOGIA','ZOOTECNIA','GANAD'],
           'TECH':['TECNOLOGIA SISTEMAS','TECNOLOGIA ELECTRICIDAD','TECNOLOGIA MECANICA','TECNOLOGIA INDUSTRIAL'],
           'ARTE':['ARTE','MUSIC','FOTOGRA','CINE','TV','ESCULT','JOYERO'],
           'ESTUDIANTES':['ESTUDI']}

with pd.option_context('mode.chained_assignment', None):  #Ignore warnings
    # Create clusters of professions
    df['profesion_cat'] = df.loc[:,'prof_ocup'].values

    # Cluster all engeneerings, designers, administrators
    df.loc[df.profesion_cat.str.find('INGEN') != -1,'profesion_cat'] = 'INGENIERIA'
    df.loc[df.profesion_cat.str.find('DISEÑ') != -1,'profesion_cat'] = 'DISEÑO'
    df.loc[df.profesion_cat.str.find('ADMIN') != -1,'profesion_cat'] = 'ADMINIST'
    
    for categ in replace.keys():
        for i in replace[categ]:
            df.loc[df.profesion_cat.str.find(i) != -1,'profesion_cat'] = categ

    # NaNs
    df.loc[df.profesion_cat.str.find('NAN ') != -1,'profesion_cat'] = np.nan
    
    # Everything else
    new_cats = ['ARTE','TECNOLOG','AGRO','SOCIAL','CONSTRU','ECONOM','SALUD',
                'EDUCACION','INGENIERIA','DISEÑO','ADMINIST']
    
    df.loc[(~df.profesion_cat.isin(new_cats)) & (~df.profesion_cat.isna()),'profesion_cat'] = 'OTROS'
    
print()
print(df.profesion_cat.value_counts(dropna=False).head(50))

# Check if profesion_cat is imbalanced as well
plotFeat_seg(df,'profesion_cat',lim=30,rot=90)
```

```python
def MAPE(y_pred,y_true):
    y_true = y_true.get_label()
    y_pred = np.array(y_pred)
    
    return 'MAPE', np.mean(np.abs((y_true - y_pred)/y_true)) * 100

def mape_obj1(preds, dtrain):
    labels = dtrain.get_label()
    grad = 100*(preds-labels)/(1.25+labels*np.abs(preds-labels))
    hess = 1e-1*np.ones(len(preds));
    return grad,hess

    
def xgtrain(X_train,X_valid,y_train,y_valid,Nestim=5,lr=0.01):
    XGBParams = {'objective' : 'reg:squarederror','nestim':Nestim,
                 "booster":"gbtree",'eta': lr,'max_depth': 20,
                 'subsample': 1.0,"alpha": 0.0,"lambda": 0.0,"gamma":0.0,'nthread':4,
                 'disable_default_eval_metric': 1,'seed':1}
    
    NRounds = 500

    dtrain = xgb.DMatrix(X_train, label=y_train)
    dvalid = xgb.DMatrix(X_valid, label=y_valid)

    xg1 = xgb.train(XGBParams, dtrain, NRounds, 
                    verbose_eval=True,#obj=mape_obj1,          
                    feval = MAPE,
                    evals=[(dtrain, 'dtrain'), (dvalid, 'dvalid')],)
    
    return xg1
  
reg = xgtrain(X_train,X_valid,y_train,y_valid,5000,0.005)
```

```python
df = (pd.read_csv('../csvs/Dataton_train_semicolon.csv',nrows=200000,
                 names=rows,sep=';')
      .replace('\\N',np.nan))

df.gasto_familiar = df.gasto_familiar.astype(float)

# Drop nans in response variable
# df = df[~df.gasto_familiar.isna()]

# For some cities with Ñ in the name: sometimes it is changed to an #, so handle that
location = ['ciudad_residencia','departamento_residencia','ciudad_laboral','departamento_laboral']
df[location] = df[location].apply(lambda x: x.str.replace('#','Ñ'))

df.head()
```

## Let's see, for the same client, how much their spendings vary with time.

```python
df.groupby('id_cli').gasto_familiar.describe()
```

### We see values are quite disperse for a single client!

Taking into account the nature of the evaluation metric, we may select one specific value out of all the possible values for a single client and use that for training so that in our training DS a user appears only once. 

In that case, as MAPE is much more sensitive to low values of y, we may want to select for instance the 25th percentile. 

# Evaluate this

With this, we'd reduce our DS from ~11M rows to only 1.1M while exploiting the various rows each client has!

```python
quant_list = [0,0.01,0.05,0.1,0.2,0.3]
a = df[df.gasto_familiar!=0].copy()
for i,p in enumerate(quant_list):

    grp = (a.groupby('id_cli')
           .gasto_familiar
           .quantile(p)
           .reset_index()
           .rename(columns={'gasto_familiar':f'gasto_familiar_perc_{i}'}))

    a = a.merge(grp,how='left',on='id_cli')
    perc_zeros = (a[f'gasto_familiar_perc_{i}'] == 0).sum()/a.shape[0]
    a = a[a.gasto_familiar != 0]
    mape = MAPE(a.gasto_familiar,a[f'gasto_familiar_perc_{i}'])
    
    print(f"Quant {p:.3f}, MAPE: {mape:.3f}, % of zeros: {perc_zeros:.3f}")
```

### This approach is not good enough. Even using the best constant value for each client, we still get a MAPE > 72, meaning any prediction for new clients will have a larger error.
